## 03. 텐서플로워 연산
 * 그래프의 생성과 그래프의 실행

### 3-1 그래프의 생성
 * 그래프는 간단하게 말하면 텐서들의 연산 모음.
 * 텐서와 텐서의 연산들을 먼저 정의하여 그래프를 만든다.

```
a = tf.constant(10)
b = tf.constant(25)
c = tf.add(a,b)
print(c )
print(type(c))
```

### 3-2 그래프의 실행
 * 실제 연산 부분 C++로 구현한 코어 라이브러리에서 수행
 * 모델 구성과 실행을 분리하여 프로그램을 깔끔하게 작성

```
sess = tf.Session()
print(sess.run(hello))
print(sess.run([a,b,c]))

sess.close()
```

### 3-3 플레이스홀더와 변수
 * 플레이스 홀더(Placeholder) : 나중에 값을 입력받기 위한 사용되는 매개변수(parameter)
 * shape(?,3)은 두번째 차원 요소가 3임을 의미한다.

### 플레이스 홀더 선언

```
# None은 크기가 정해져 있지 않음을 의미
X = tf.placeholder(tf.float32, [None, 3])
print(X)
```

### 추후에 넣을 데이터 x

```
x_data = [[1,2,3], [4,5,6]]
```

### 변수 정의
 * tf.random_normal 함수를 이용한 정규 분포의 무작위 행렬 3,2 생성

```
W = tf.Variable(tf.random_normal([3,2]))
b = tf.Variable(tf.random_normal([2,1]))
print(W)
print(b)
```

### 수식 계산


```
# X = tf.placeholder(tf.float32, [None, 3])   # [? * 3] X [3 * 2] => ? X 2 
# W = tf.Variable(tf.random_normal([3,2]))

expr = tf.matmul(X, W) + b
expr
```

### 연산 실행 및 결과 출력

```
sess = tf.Session()
sess.run(tf.global_variables_initializer())

print("=== x_data ===")
print(x_data)
print("=== W matrix ===")
print(sess.run(W))
print("=== b ===")
print(sess.run(b)) 
print("=== expr ===")
print(sess.run(expr, feed_dict={X:x_data}))

sess.close()
```
